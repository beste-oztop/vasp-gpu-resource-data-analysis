```python
import pandas as pd

import sys
sys.path.append('../scripts')

from metric_analysis import load_large_query_csv, create_combined_df
from plot_functions import plot_pie_chart
from code_group import match_name, match_algo
```

```python
file_path1 = '/global/homes/b/boztop/dcgm_slurm_studies/slurm_data/perlmutter_gpu_jobs_march_2025.csv'
file_path2 = '/global/homes/b/boztop/dcgm_slurm_studies/slurm_march_gpu_jobs.csv'

```

```python
march_vasp_df = pd.read_csv(file_path1, low_memory=False)
march_vasp_df= march_vasp_df[~(march_vasp_df["NodeList"].str.contains("login", case=False, na=False) |
march_vasp_df["QOS"].str.contains("cron", case=False, na=False) |
march_vasp_df["QOS"].str.contains("gpu_debug", case=False, na=False) |
march_vasp_df["State"].str.contains ("RUNNING", case=False, na=False)
)]
```

```python
march_slurm_df = load_large_query_csv(file_path2)
plot_pie_chart(march_slurm_df)
```

```python
march_slurm_df = march_slurm_df[march_slurm_df['JobID'].str.endswith('.0')]
march_slurm_df['JobID'] = march_slurm_df['JobID'].str.replace(r'\.0$', '', regex=True)

set1 = set(march_vasp_df['JobID'])
set2 = set(march_slurm_df['JobID'])

matches = set1.intersection(set2)
not_matches = set1.difference(set2)
print(f"Number of matching JobIDs: {len(matches)}")

```

```python
df_merged = march_vasp_df.merge(march_slurm_df[['JobID', 'exename']], on='JobID', how='left')
df_merged['CodeGroup'] = df_merged['exename'].apply(match_name)
```

```python
vasp_job_df = df_merged[df_merged['CodeGroup'] == 'VASP']
vasp_job_df = vasp_job_df.dropna(axis=1, how='all')
vasp_job_df = vasp_job_df[vasp_job_df['QOS'].isin(['gpu_regular', 'gpu_premium'])]

vasp_job_df.to_csv('vasp_job_df.csv', index=False)
```
