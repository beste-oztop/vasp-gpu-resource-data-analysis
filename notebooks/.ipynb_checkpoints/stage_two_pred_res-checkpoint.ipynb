Confusion Matrices (Using Results from Runtime Power Prediction)

```python
# VASP - max
cm_max = np.array([
    [21403, 11855,  2087,   944],
    [ 3043, 91929, 31017, 28135],
    [  177,  7828, 27891, 30768],
    [   90,  5028,  8011, 78877]
])

# VASP - mean
cm_mean = np.array([
    [27238,  8269,   631,   151],
    [11303,108948, 24185,  9688],
    [  326, 14912, 34800, 16626],
    [  171, 11853, 30513, 49469]
])


# VASP - lgb
cm_lgb = np.array([
    [32712,  3259,   194,   124],
    [ 2397,136368,  9681,  5681],
    [  290, 13884, 36825, 15665],
    [   54,  3942,  8584, 79426]
])


cms = [cm_max, cm_mean, cm_lgb]
titles = ['Max Baseline', 'Mean Baseline', 'Proposed Framework']

cms_norm = [cm.astype(float) / cm.sum(axis=1, keepdims=True) for cm in cms]

label_names = ['51100W', '100150W', '150200W', '>200W']
cmaps = ['Greens', 'Blues', 'Reds']

fig, axes = plt.subplots(1, 3, figsize=(15, 8))

for ax, cm_norm, title, cmap in zip(axes, cms_norm, titles, cmaps):
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm_norm,
        display_labels=label_names
    )

    disp.plot(
        ax=ax,
        cmap=cmap,
        values_format=".2f",
        colorbar=False,
        text_kw={'fontsize': 14, 'fontweight': 'bold'}
    )

    im = ax.imshow(cm_norm, cmap=cmap, vmin=0, vmax=1)
    
    cbar = plt.colorbar(
        im,
        ax=ax,
        fraction=0.045,  
        pad=0.04 
    )
    cbar.ax.tick_params(labelsize=14)

    ax.set_title(title, fontsize=22)
    ax.set_xlabel('Predicted label', fontsize=18)
    ax.set_ylabel('True label', fontsize=18)
    ax.tick_params(axis='both', labelsize=14)

    for label in ax.get_xticklabels():
        label.set_rotation(45)

plt.tight_layout()
plt.savefig("cm_stage2_vasp.pdf", format="pdf", bbox_inches="tight")
plt.show()

```

```python
y_true_lgb  = lgb_model_res['y_test']
y_true_mean = mean_model_res['y_test']
y_true_max  = max_model_res['y_test']

y_pred_lgb  = lgb_model_res['y_pred']
y_pred_mean = mean_model_res['y_pred']
y_pred_max  = max_model_res['y_pred']

print(len(y_pred_mean))
print(len(y_pred_max))
print(len(y_pred_lgb))
print(" .. ")

n_points = 60
start_idx = 3600 # Can be anything, really.. I chose this part since the dynamic changes in power utilization is visible!
end_idx   = start_idx + n_points

y_true_plot  = y_true_lgb[start_idx:end_idx]
# scale 50->250
y_lgb_plot   = y_pred_lgb[start_idx:end_idx] * 50 + 55 
#if beyond 200, ceil to 220
y_lgb_plot = np.where(y_lgb_plot > 200, 220, y_lgb_plot)
y_mean_plot  = y_pred_mean[start_idx:end_idx] * 50 + 55 
y_mean_plot = np.where(y_mean_plot > 200, 220, y_mean_plot)
y_max_plot   = y_pred_max[start_idx:end_idx] * 50 + 55 
y_max_plot = np.where(y_max_plot > 200, 220, y_max_plot)

x = np.arange(n_points)

plt.figure(figsize=(16,5))
plt.plot(x, y_true_plot, label='Actual Utilization', color='black', linewidth=2, marker='o', markersize=5)

plt.step(x, y_max_plot, label='Max Baseline', color='seagreen', linestyle='--', where='mid', linewidth=3, alpha=0.8)
plt.step(x, y_mean_plot, label='Mean Baseline', color='cornflowerblue', linestyle='--', where='mid', linewidth=3, alpha=0.8)
plt.step(x, y_lgb_plot, label='Proposed Framework', color='crimson', linestyle='-', where='mid', linewidth=3, alpha=0.8)

# plt.step(x, y_max_plot, label='Max Baseline', color='seagreen', where='mid', linewidth=3)
# plt.step(x, y_mean_plot, label='Mean Baseline', color='cornflowerblue',  where='mid', linewidth=3)
# plt.step(x, y_lgb_plot, label='Proposed Framework', color='crimson', where='mid', linewidth=3)


plt.ylabel('Average Power Utilization (W) \n (classes are scaled  to 55-220)', fontsize=20)
plt.tick_params(axis='both', labelsize=16)
plt.legend(fontsize=14,loc='best')
plt.grid(axis='both',color='gray', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.savefig("example_run.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 4))

# Convert to numpy arrays to avoid dimension issues
y_true_plot_arr = np.array(y_true_plot).flatten()
y_max_plot_arr = np.array(y_max_plot).flatten()
y_mean_plot_arr = np.array(y_mean_plot).flatten()
y_lgb_plot_arr = np.array(y_lgb_plot).flatten()

models = [
    ('Max Baseline', y_max_plot_arr, 'seagreen'),
    ('Mean Baseline', y_mean_plot_arr, 'cornflowerblue'),
    ('Proposed Framework', y_lgb_plot_arr, 'crimson')
]

class_boundaries = [0, 100, 150, 200, float('inf')]
def get_class(value):
    for i in range(len(class_boundaries) - 1):
        if class_boundaries[i] <= value < class_boundaries[i + 1]:
            return i
    return len(class_boundaries) - 2


for idx, (name, y_pred, color) in enumerate(models):
    ax = axes[idx]
    
    # Plot actual utilization
    ax.plot(x, y_true_plot_arr, label='Actual Utilization', color='black', linewidth=2.5, zorder=5)
    
    # Plot model prediction
    ax.step(x, y_pred, label=name, color=color, where='mid', linewidth=2.5, zorder=4)
    
    # Shade only where prediction is in wrong class
    for i in range(len(x)):
        true_class = get_class(y_true_plot_arr[i])
        pred_class = get_class(y_pred[i])
        
        if true_class != pred_class:
            ax.fill_between([x[i], x[i] + 0.5], y_true_plot_arr[i], y_pred[i], 
                           alpha=0.3, color=color, zorder=2)
    if idx ==0:        
        ax.set_ylabel('Average Power Utilization (W)', fontsize=16)
    # ax.set_xlabel('Time Points', fontsize=14)
    plt.tick_params(axis='both', labelsize=16)
    # ax.set_title(name, fontsize=20)
    ax.legend(fontsize=16, loc='lower center')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.savefig("example_run_other.pdf", format='pdf', bbox_inches='tight', dpi=300)
plt.show()
```
